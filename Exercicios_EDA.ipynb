### Descrição dos dados

O dataset possui 891 registros (linhas) e 12 atributos (colunas).

Abaixo estão as descrições das colunas do nosso Dataset:

* passengerid = ID do passageiro do navio (código primário).
* survived = Se o passageiro sobreviveu ao naufrágio estará como 1, caso contrário estará com 0.
* pclass = Tipo de classe de passagem (do 1 ao 3), sendo 1 a melhor classe.
* name = Nome do passageiro.
* sex = Gênero do passageiro, sendo masculino e feminino.
* age = Idade do passageiro na data da ocorrência do naufrágio.
* sibsp = Número de irmãos / cônjuges a bordo.
* parch = Número de pais / filhos a bordo.
* ticket = Código do ticket.
* fare = Valor da passagem.
* cabin = Código de identificação da Cabine.
* embarked = Local ondem o passageiro embarcou no navio.

#### Carregando os dados

```python
import pandas as pd


# Carregando os dados
# O arquivo “titanic.csv” deve estar salvo na mesma pasta do seu código Python
dados = pd.read_csv("titanic.csv")

# Exibe o número de linhas e colunas
print(dados.shape)

# Exibe as 10 primeiras linhas
dados.head()
```

#### 1. Qual foi a proporção de passageiros sobreviventes?

Dica: use a função value_counts() do pandas com o parâmetro _normalize_

```python
# Seu código
print(round(dados["Survived"].value_counts(normalize=True)[1]*100, 2),"% sobreviventes")
dados["Survived"].value_counts(normalize=True)
```

#### 2. Qual era a faixa etária dos passageiros que estavam no Titanic?

```python
# Seu código
print("De ",dados["Age"].min()," a ",dados["Age"].max())
```

#### 3. Qual porcentagem de homens e mulheres sobreviventes?

```python
# Seu código
#************************
print("Mulheres: ",dados["Sex"].loc[dados["Survived"] == 1].value_counts(normalize=True)["female"])
print("Homens: ",dados["Sex"].loc[dados["Survived"] == 1].value_counts(normalize=True)["male"])
```

#### 8. Cria um histograma para os atributos age, pclass e fare. Quais informações você pode extrair desses gráficos?


```python
# Seu código
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, tight_layout=True, figsize=(10,8))
axs[0][0].set_title("Age")
axs[0][0].hist(dados["Age"])
axs[0][0].grid()
axs[0][1].set_title("Fare")
axs[0][1].hist(dados["Fare"])
axs[0][1].grid()
axs[1][0].set_title("Pclass")
axs[1][0].hist(dados["Pclass"])
axs[1][0].grid()

print("A maioria dos tripulantes tinha entre 20 e 30 anos.")
print("Havia uma quantidade mínima de tripulantes com mais de 70 anos.")
print("Grande parte dos tripulantes paragam menos de 100.")
print("Um pequeno grupo pagou 500.")
print("A maioria dos tripulantes eram da terceira classe.")
print("Haviam mais tripulantes de primeira classe do que segunda.")

```
